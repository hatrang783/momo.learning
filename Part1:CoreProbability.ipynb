##Permutation of Distinct Object

An ordered arrangement of n distinct object

```python
def factorial(x):
  ret = 1
  for i in range(2,x+1):
    ret *= i
  return ret

factorial(3)
```

```python
#Number of unique orderings of 5 characters: "B", "A", "Y", "E", "S":
result = factorial(5)
result
```

##Combinations of Distinct Objects
an unordered selection of r objects from a set of n objects

```python
# Ways to select 3 from 5 unique items

result = (5*4*3) / factorial(3)
r_use_formulas = factorial(5) / (factorial(3) * factorial(5-3))

print(result, r_use_formulas)
print(set(itertools.combinations([1,2,3,4,5], 3)))
```

```python
# In the Hunger Games, how many ways are there of choosing 2 villagers from district 12,
# which has a population of 8,000?

result = factorial(8000) / (factorial(2)*factorial(8000-2))
result
```

```python
def combination(n, r):
  return factorial(n) / (factorial(r) * factorial(n-r))

combination(4,2)
```

#Probability of AND

```python
def prob_and(e_probs=[], independence=True):
  if independence:
    ret = 1
    for i in e_probs:
      ret *= i
    return ret

prob_and([0.1, 0.2, 0.15])
```
